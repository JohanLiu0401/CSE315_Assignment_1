```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')#使用ggplot样式
%matplotlib inline

def dropMisVal(df):
    df_copy = df.dropna() 
    return df_copy   #这里并没有对原数据进行修改,而是生成一个新的移除缺省值行的数据帧

def handMisVal(df,value):
    if value == 'median':
        df_copy = df.fillna(df.median())
    elif value == 'mean':
        df_copy = df.fillna(df.mean())
    elif value == 'mode':
        df_copy = df.fillna(df.mode().iloc[0])  #某些属性不只有一个众数，所以选择使用第一组

    return df_copy


iris = pd.read_csv('irismissing.csv', header=0, usecols=[2,3,4,5,6])    #usecols属性设置读取的指定列


#统计缺省值行数
print ("The row number of the instances that have missing values: ")
null_row = 0
for i in range(0,len(iris)):
    if iris.iloc[i,:].count() != 5 :
        print (i)
        null_row += 1
print ("Total:", null_row)



#清除具有缺省值的行
df_drop = dropMisVal(iris)
df_median = handMisVal(iris,'median')
df_mean = handMisVal(iris,'mean')
df_mode = handMisVal(iris,'mode')

#作图

#条形图
df_drop.plot.bar()
#直方图
df_drop.plot.hist(alpha=0.5)
#箱形图
df_drop.plot.box()
#区域图
df_drop.plot.area()

```
