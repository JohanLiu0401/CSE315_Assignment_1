```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')#ggplot
%matplotlib inline

def dropMisVal(df):
    df_copy = df.dropna() 
    return df_copy   #,

def handMisVal(df,value):
    if value == 'median':
        df_copy = df.fillna(df.median())
    elif value == 'mean':
        df_copy = df.fillna(df.mean())
    elif value == 'mode':
        df_copy = df.fillna(df.mode().iloc[0])  #

    return df_copy


iris = pd.read_csv('irismissing.csv', header=0, usecols=[2,3,4,5,6])    #usecols


#
print ("The row number of the instances that have missing values: ")
null_row = 0
for i in range(0,len(iris)):
    if iris.iloc[i,:].count() != 5 :
        print (i)
        null_row += 1
print ("Total:", null_row)



#
df_drop = dropMisVal(iris)
df_median = handMisVal(iris,'median')
df_mean = handMisVal(iris,'mean')
df_mode = handMisVal(iris,'mode')

#

#
df_drop.plot.bar()
#
df_drop.plot.hist(alpha=0.5)
#
df_drop.plot.box()
#
df_drop.plot.area()

```
